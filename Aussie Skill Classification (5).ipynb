##### CREATING SYNTHETIC DATASET DUE TO LACK OF DATA IN ACTUAL CORE COMPETENCIES DATASET

```python
# Load the Aussie dataset
df = pd.read_csv('Aussie dataset.csv')

# Extract unique job titles from the original dataset
job_titles = df['ANZSCO_Title'].unique()

# Function to generate synthetic data with slight variations
def generate_synthetic_data():
    synthetic_data = []
    
    # Define a small range of noise (e.g., -1 to 1) to add variability
    noise_range = (-1, 1)
    
    for job_title in job_titles:
        # Filter the Aussie dataset to get rows matching the selected job title
        matching_jobs = df[df['ANZSCO_Title'] == job_title]
        
        for _ in range(5):  
            # Randomly select scores and add small random noise
            digital_engagement = round(random.choice(matching_jobs['Digital engagement'].values) + random.uniform(*noise_range))
            initiative_innovation = round(random.choice(matching_jobs['Initiative and innovation'].values) + random.uniform(*noise_range))
            learning = round(random.choice(matching_jobs['Learning'].values) + random.uniform(*noise_range))
            numeracy = round(random.choice(matching_jobs['Numeracy'].values) + random.uniform(*noise_range))
            oral_communication = round(random.choice(matching_jobs['Oral communication'].values) + random.uniform(*noise_range))
            planning_organising = round(random.choice(matching_jobs['Planning and organising'].values) + random.uniform(*noise_range))
            problem_solving = round(random.choice(matching_jobs['Problem solving'].values) + random.uniform(*noise_range))
            reading = round(random.choice(matching_jobs['Reading'].values) + random.uniform(*noise_range))
            teamwork = round(random.choice(matching_jobs['Teamwork'].values) + random.uniform(*noise_range))
            writing = round(random.choice(matching_jobs['Writing'].values) + random.uniform(*noise_range))
            
            # Ensure that scores stay within valid bounds (e.g., 1 to 10)
            def clamp(value, min_value=1, max_value=10):
                return max(min_value, min(value, max_value))
            
            digital_engagement = clamp(digital_engagement)
            initiative_innovation = clamp(initiative_innovation)
            learning = clamp(learning)
            numeracy = clamp(numeracy)
            oral_communication = clamp(oral_communication)
            planning_organising = clamp(planning_organising)
            problem_solving = clamp(problem_solving)
            reading = clamp(reading)
            teamwork = clamp(teamwork)
            writing = clamp(writing)
            
            # Append the synthetic row
            synthetic_data.append([ 
                job_title, digital_engagement, initiative_innovation, learning,
                numeracy, oral_communication, planning_organising, problem_solving,
                reading, teamwork, writing
            ])
    
    # Create the DataFrame with correct column names
    synthetic_df = pd.DataFrame(synthetic_data, columns=[
        'ANZSCO_Title', 'Digital engagement', 'Initiative and innovation', 'Learning',
        'Numeracy', 'Oral communication', 'Planning and organising', 
        'Problem solving', 'Reading', 'Teamwork', 'Writing'
    ])
    
    return synthetic_df

# Generate the synthetic dataset
synthetic_dataset = generate_synthetic_data()

# Save the synthetic dataset to a CSV file
synthetic_dataset.to_csv('synthetic_dataset_with_noise.csv', index=False)

# Display the synthetic dataset
synthetic_dataset.head()
```

### RANDOM FOREST

```python

# Initialize the Random Forest classifier
rf = RandomForestClassifier()

# Fit the Random Forest model on the training data
rf.fit(X, y)

# Predict the classes
y_pred_rf = rf.predict(X_test)

# Calculate the accuracy of the classifier
accuracy = accuracy_score(y_test, y_pred_rf)

# Print the accuracy
print("Accuracy: {:.2f}%".format(accuracy * 100))
```

```python
# Predict class probabilities
probabilities = rf.predict_proba(X_test)

# Define a function to get the top N suited classes
def get_top_n_classes(probabilities, n=10):
    top_n_classes = []
    for probs in probabilities:
        # Get indices of the top N probabilities
        top_indices = np.argsort(probs)[::-1][:n]
        top_n_classes.append(top_indices)
    return top_n_classes

# Get top 5 suited classes for each instance
top_5_classes = get_top_n_classes(probabilities, n=10)

# Example of output
print("Top 5 suited classes for the first test instance:")
print(top_5_classes[0])
print("Class names:")
print([rf.classes_[i] for i in top_5_classes[69]])
```

```python
def get_top_job_titles(pipeline, X, N=3):
    # Predict class probabilities
    probabilities = pipeline.predict_proba(X)
    
    # Get the indices of the top N probabilities
    top_indices = np.argsort(probabilities[0])[-N:][::-1]
    
    # Get the job titles corresponding to these indices
    predicted_job_titles = [pipeline.classes_[i] for i in top_indices]
    
    return predicted_job_titles
```
